### matplotlib

#### 图形绘制

```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
x = np.linspace(0, 2*np.pi,100)
y = np.sin(x)
plt.plot(x, y)
# 设置坐标的范围
plt.xlim(-1, 10)
plt.ylim(-1.5, 1.5) 
# grid网格线 alpha:透明度
plt.grid(color='green', alpha=0.2, linestyle='--')
```
