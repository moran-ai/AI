```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi,20)
y1 = np.sin(x)
y2 = np.cos(x)

# 设置颜色，线型，点型
plt.plot(x,y1,color = 'indigo',ls = '-.',marker = 'p')
plt.plot(x,y2,color = '#FF00EE',ls = '--',marker = 'o')
plt.plot(x,y1 + y2,color = (0.2,0.7,0.2),marker = '*',ls = ':')
plt.plot(x,y1 + 2*y2,linewidth = 3,alpha = 0.7,color = 'orange') # 线宽、透明度
# 参数连用 b --> blue  o  marker 圆圈  -- 线的样式 虚线
plt.plot(x,2*y1 - y2,'bo--') # 参数连用
```

#### 多图布局

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi)
# sharex:所有小图共享x轴  sharey:表示所有小图共享y轴  坐标轴以所有小图中范围最大的进行显示
fig, ((ax11,ax12,ax13), (ax21,ax22,ax23),(ax31,ax32,ax33)) = plt.subplots(3, 3)
# 也可通过plt.subplot() 一个个添加子视图
fig.set_figwidth(9) # 画布的宽度
fig.set_figheight(6) # 画布的高度
# 下面画9个子视图
ax11.plot(x,np.sin(x))
ax12.plot(x,np.cos(x))
ax13.plot(x,np.tanh(x))
ax21.plot(x,np.tan(x))
ax22.plot(x,np.cosh(x))
ax23.plot(x,np.sinh(x))
ax31.plot(x,np.sin(x) + np.cos(x))
ax32.plot(x,np.sin(x*x) + np.cos(x*x))
ax33.plot(x,np.sin(x)*np.cos(x))
# 紧凑显示，边框会比较小，可以注释掉该行查看效果
plt.tight_layout()
plt.show()
```

### 文本,箭头,注释

#### 文本

```python
import numpy as np
import matplotlib.pyplot as plt

# 字体属性
font = {'fontsize': 20,
        'family': 'KaiTi',
        'color':  'red',
        'weight': 'bold'}

x = np.linspace(0.0, 5.0, 100)
y = np.cos(2*np.pi*x) * np.exp(-x)
plt.rcParams['axes.unicode_minus'] = False  # 正常显示负号
plt.figure(figsize=(9,6))
# plt.rcParams['font.family'] = 'KaiTi'
plt.plot(x, y, 'k')
# pad：标题和图形之间的距离
# 子视图小标题
plt.title('exponential decay',fontdict=font, pad=20)
# 整个的大标题
plt.suptitle('指数衰减',y = 1.05,fontdict = font,fontsize = 30)
# 设置文本内容  x,y为位置
plt.text(x = 2, y = 0.65, # 横纵坐标位置(2, 0.65)
         s = r'$\cos(2 \pi t) \exp(-t)$',fontdict=font) # 文本内容
plt.text(x = 2, y = 0.4, s = 'Hello Word!')
plt.xlabel('time (s)')  # 设置x轴标签
plt.ylabel('voltage (mV)')  # 设置y轴标签
plt.show()
```

#### 箭头

```python
import matplotlib.pyplot as plt
import numpy as np
loc = np.random.randint(0,10,size = (10,2))

plt.figure(figsize=(10, 10))
# 获取所有的行，然后0是第一列，作为横坐标  1是第二列，作为纵坐标 ms为大小
plt.plot(loc[:,0], loc[:,1], 'g*', ms=20)
plt.grid(True)  # 网格线

# 路径
way = np.arange(10)
np.random.shuffle(way)  # 对数据进行打乱

# 10个点，连接9个点 最后一个点不需要连接
for i in range(0, len(way)-1):
    start = loc[way[i]]  # 起点
    end = loc[way[i+1]]   # 终点
    # 画箭头 arrow
    # start[0] start[1]:开始的位置（x, y）   end[0]-start[0]：水平坐标 end[1] -start[1]：垂直坐标
    plt.arrow(start[0], start[1], end[0]-start[0], end[1]-start[1], # 坐标与距离
              head_width=0.2, lw=2,#箭头长度，箭尾线宽
              length_includes_head = True) # 长度计算包含箭头箭尾
    plt.text(start[0],start[1],s = i,fontsize = 18,color = 'red') # 文本
    if i == len(way) - 2: # 最后一个点
        plt.text(end[0],end[1],s = i + 1,fontsize = 18,color = 'blue')
        start = loc[way[-1]]  # 开始的位置，为最后一个点
        end = loc[way[0]]  # 结束的位置，为第一个点
        # 画箭头
        plt.arrow(start[0], start[1], end[0]-start[0], end[1]-start[1],
                 head_width=0.2, lw=2, length_includes_head=True)
        # 文本内容
        plt.text(start[0], start[1], s = i, fontsize=18, color='blue')
```

#### 注释 (箭头+文本)

```python
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x = np.arange(0.0, 5.0, 0.01)
y = np.cos(2*np.pi*x)
line, = ax.plot(x,y,lw=2)
ax.annotate('local max', # 文本内容
            xy=(2, 1),  # 箭头指向位置
            xytext=(3, 1.5), # 文本位置
            arrowprops=dict(facecolor='black', shrink=0.05)) # 箭头
ax.annotate('local min',
            xy = (2.5,-1),
            xytext = (4,-1.8),
            arrowprops = dict(facecolor = 'black',
                              width = 2, # 箭头宽度
                              headwidth = 10,# 箭头头部宽度
                              headlength = 10, # 箭头头部长度
                              shrink = 0.1)) # 箭头两端收缩的百分比（占总长）
ax.annotate('median',
            xy = (2.25,0),
            xytext = (0.5,-1.8),
            arrowprops = dict(arrowstyle = '-|>'), # 箭头样式
            fontsize = 20)
ax.set_ylim(-2, 2)
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# 标签
vegetables = ["cucumber", "tomato", "lettuce", "asparagus","potato", "wheat", "barley"]
farmers = list('ABCDEFG')  # 农民个数

# 随机数 正态分布
harvest = np.random.randn(7,7)*5 # 农民丰收数据

plt.rcParams['font.size'] = 18
plt.rcParams['font.weight'] = 'heavy'
plt.figure(figsize=(9,9))
im = plt.imshow(harvest, cmap='autumn')  # cmap:颜色控制

# 绘制文本
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = plt.text(j, i, round(harvest[i, j],1),
                       ha="center", va="center", color='r')
plt.title("Matplotlib bakery: A donut",fontsize = 18,pad = 25)
plt.tight_layout() # 紧凑显示
```

#### 面积图

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,6))
days = [1,2,3,4,5]  
sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]   
plt.stackplot(days,sleeping,eating,working,playing)  
plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot',fontsize = 18)
plt.legend(['Sleeping','Eating','Working','Playing'],fontsize = 18)
```

#### 蜘蛛图

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'KaiTi'
labels=np.array(["个人能力","IQ","服务意识","团队精神","解决问题能力","持续学习"])
stats=[83, 61, 95, 67, 76, 88] # y
# 画图数据准备，角度、状态值
angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)  # x
stats=np.concatenate((stats,[stats[0]]))  #首尾相接  stats最后一个数据 stats[0]:最后一个数据
angles=np.concatenate((angles,[angles[0]]))  # 首尾相接
# 用Matplotlib画蜘蛛图
fig = plt.figure(figsize=(9,9))
ax = fig.add_subplot(111, polar=True)   
ax.plot(angles, stats, 'o-', linewidth=2) # 连线
ax.fill(angles, stats, alpha=0.25) # 填充
# 设置角度
ax.set_thetagrids(angles[:-1]*180/np.pi,#角度值
                  labels,fontsize = 18)
_ = ax.set_rgrids([20,40,60,80],fontsize = 18)
```

### 3D图形

#### 3D散点图和3D图

```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d.axes3d import Axes3D

x = np.linspace(0, 60, 300)
y = np.sin(x)
z = np.cos(x)
plt.rcParams['axes.unicode_minus'] = False
# 画3D图形
# fig = plt.figure(figsize=(9, 6))
# ax = Axes3D(fig)
# ax.plot(x, y, z)
plt.figure(figsize=(9, 6))
ax = plt.subplot(111, projection='3d')
ax.plot(x, y, z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# 散点图
x_ = np.random.randint(0, 20, size=20)
y_ = np.random.randn(20)
z_ = np.random.randn(20)
ax.scatter(x_, y_, z_, color='r')
ax.view_init(elev=30, azim=-80)  # 调整角度
```

#### 3D条形图

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')  # 忽略警告
from mpl_toolkits.mplot3d.axes3d import Axes3D # 3D引擎
month = np.arange(1,5)
# 每个月 4周 每周都会产生数据
# 三个维度：月、周、销量
fig = plt.figure(figsize=(9,6))
ax3 = Axes3D(fig)

for m in month:
    ax3.bar(np.arange(1, 5),  # 横坐标
            np.random.randint(1,10,size = 4),  # 纵坐标
            zs = m ,  # 月份
            zdir = 'x',# 在哪个方向上，一排排排列
            alpha = 0.7,# alpha 透明度
            width = 0.5)
ax3.set_xlabel('月份',fontsize = 18,color = 'red')
ax3.set_xticks(month)
ax3.set_ylabel('周',fontsize = 18,color = 'red')
ax3.set_yticks([1,2,3,4])
ax3.set_zlabel('销量',fontsize = 18,color = 'green')
```

### seaborn

```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

# 快速设置主题
sns.set(style='dark', context='poster', font='KaiTi')
plt.figure(figsize=(9, 6))

x = np.linspace(0, 2*np.pi, 20)
y = np.sin(x)

sns.lineplot(x = x, y = y, color='green', ls = '-')
sns.lineplot(x = x, y = np.cos(x), color='red', ls = '-')
```

#### 线型图

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set(style = 'dark',context = 'notebook',font = 'STKaiti') # 设置样式
plt.figure(figsize=(9,6))
fmri = pd.read_csv('./fmri.csv') # fmri这一核磁共振数据
ax = sns.lineplot(x = 'timepoint',y = 'signal',
hue = 'event', # 根据event属性分类绘制
                  style = 'event' ,
data= fmri,
palette='deep',  # 画板，颜色
markers=True,
markersize = 10)
plt.xlabel('时间节点',fontsize = 30)
# plt.savefig('./线形图.png',dpi = 200)
```

#### 热力图

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,9))
flights = pd.read_csv('./flights.csv')
# pivot 数据重塑，改变原来的数据结构  month为行索引   year为列索引 后面的为数据
flights = flights.pivot("month", "year", "passengers")
# annot:注释  fmt=d  整数  cmap:颜色
sns.heatmap(flights, annot=True,fmt = 'd',cmap = 'RdBu_r',
linewidths=0.5)
```
