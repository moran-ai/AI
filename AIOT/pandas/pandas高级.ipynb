#### 方式一：concat

#### 方式三 join SQL 数据库风格数据操作 merge 



```python
import pandas as pd
import numpy as np
# 表一中记录的是name和体重信息
df1 = pd.DataFrame(data = {'name':['softpo','Daniel','Brandon','Ella'],'weight':[70,55,75,65]})
# 表二中记录的是name和身高信息
df2 = pd.DataFrame(data = {'name':['softpo','Daniel','Brandon','Cindy'],'height':[172,170,170,166]})
df3 = pd.DataFrame(data = {'名字':['softpo','Daniel','Brandon','Cindy'],'height':[172,170,170,166]})
# display(df1, df2, df3)
pd.merge(df1, df2)  # 数据合并 根据共同属进行合并 
```

```python
# 指定合并的列
pd.merge(df1, df3, left_on='name', right_on='名字')
```

```python
df4 = pd.DataFrame(data = np.random.randint(0,151,size = (10,3)),
                   index = list('ABCDEFHIJK'),
                   columns=['Python','Keras','Tensorflow'])
s = df4.mean(axis=1).round(1)  # 每个人的平均分 返回的数据结构是一个Series

# 合并df4和df5 需要将Series转为DataFrame
df5 = pd.DataFrame(s, columns=['平均值'])

# 合并df4和df5  df4和df5没有相同的属性，但是行索引相同 可以对索引进行合并
pd.merge(df4, df5, left_index=True, right_index=True)
pd.concat([df4, df5], axis=1)  # 使用concat进行行索引的合并
```

### 数据转换

#### 轴和元素替换

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(data = np.random.randint(0,10,size = (10,3)),
                  index = list('ABCDEFHIJK'),
                  columns=['Python','Tensorflow','Keras'])
df.iloc[4, 2] = None
df
```

```python
# 重命名索引
df.rename(index={'A': 'AA', 'B': 'BB'}, columns={'Python': '人工智能'})
```

```python
# 替换值
df.replace(3, 1024) # 这里的替换是替换所有的数据

# 替换多个数据
df.replace([0, 7], 2034)

# 将指定的数据替换为指定的值  将0替换为34， 将NaN替换为333
df.replace({0: 34, np.nan: 333})

# 替换特定的一列数据
df.replace({'Python': 8}, -1024)
```

#### map Series 
进行映射，Series独有

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(data = np.random.randint(0,10,size = (10,3)),
                  index = list('ABCDEFHIJK'),
                  columns=['Python','Tensorflow','Keras'])
df.iloc[4, 2] = None
 
# 使用map进行映射  将1映射为Hellom, 将5映射为world, 将NaN映射为AI
df['Keras'].map({1: 'Hello', 5: 'world', np.NaN: 'AI'})

# 使用匿名函数  使用匿名函数判断x大于5还是小于5，如果大于返回True， 小于5返回False
df['Python'].map(lambda x: True if x >= 5 else False)
```

#### apply改变元素

支持Series和DataFrame

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(data = np.random.randint(0,100,size = (10,3)),
                  index = list('ABCDEFHIJK'),
                  columns=['Python','Tensorflow','Keras'])
df.iloc[4, 2] = None

# apply series数据转换
df['Keras'].apply(lambda x: True if x > 5 else False)

def covert(x):
    if x < 60:
        return '不及格'
    elif x < 85:
        return '良好'
    elif x > 85:
        return '优秀'
    
# 根据规则进行数据的转换
result=df['Python'].apply(covert)

# 将结果插入到原数据中
index = list(df.columns).index('Python') + 1
df.insert(loc=index, column='Python等级', value=result)
df
```

```python
# 进行多行的转换
import numpy as np
import pandas as pd
df = pd.DataFrame(data = np.random.randint(0,100,size = (10,3)),
                  index = list('ABCDEFHIJK'),
                  columns=['Python','Tensorflow','Keras'])
df.iloc[4, 2] = None
df

def convert(x):
    if x < 60:
        return '不及格'
    elif x < 95:
        return '良好'
    elif x > 95:
        return '优秀'

# 使用apply进行多行的转换
for col in df.columns:
    # 获取索引
    index = list(df.columns).index(col) + 1
    result = df[col].apply(convert)  # 转换结果
    # 进行转换
    df.insert(loc=index, column=col + '等级', value=result)
df
```
