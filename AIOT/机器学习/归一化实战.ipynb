```python
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('zhengqi_train.txt', sep='\t')
X_train = df.iloc[:, :-1]   # 训练数据
y_train = df[['target']]  # 目标值
y_train
```

```python
# 测试数据
X_test = pd.read_csv('zhengqi_test.txt', sep='\t')
display(X_train.head(), y_train.head(), X_test.head())
```

### 不使用归一化

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

### 使用最大值最小值进行归一化

```python
norm = MinMaxScaler()
X_train_norm = norm.fit_transform(X_train)
y_train_norm = norm.fit_transform(y_train)
X_test_norm = norm.fit_transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train_norm, y_train_norm)
```

```python
model.predict(X_test_norm)
```

### 使用0均值归一化

```python
norm = StandardScaler()
X_train_norm = norm.fit_transform(X_train)
y_train_norm = norm.fit_transform(y_train)
X_test_norm = norm.fit_transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train_norm, y_train_norm)
```

```python
model.predict(X_test_norm)[:15]
```
